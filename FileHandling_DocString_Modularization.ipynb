```python
f=open('requirements.txt','r')
data_1=f.readlines()
f.close()
f=open('requirements_new.txt','w')
#use writelines instead of write command for multiple lines
new_data=[]
for val in data_1:
    new_data.append(val.replace('@','at_the_rate'))
f.writelines(new_data)
f.close()
```
