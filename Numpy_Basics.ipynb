```python
'''
1.Create a 4,3,5 array between 30 to 240. 
2.sort the array row wise. create a list comprehension
3.Convert that list into array again
'''
a=np.random.uniform(30,240, size=(4,3,5))
print('Creating Array :')
print(a)
print(type(a))
print('Creating Sorted Array :')
print(a)
print(type(a))
np.sort(a,axis=0)
print('Creating List Comprehension :')
r=[sum(sum(inner) for inner in outersubelement) for outersubelement in a]
print(r)
print(type(r))
print('Creating Array Again :')
a=np.array(r)
print(a)
print(type(a))
```
